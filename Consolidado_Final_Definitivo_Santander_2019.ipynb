## Explotacion_Minerales

```python
# === Cargar archivo 2 ===
file_path = '/content/drive/MyDrive/Colab Notebooks/BasesDatosExcel/2.Explotacion_Minerales.xlsx'
df2 = pd.read_excel(file_path)

# Conversión segura a numérico para evitar errores
cols_num = [
    'volumenesExplotacion'
]

for col in cols_num:
    df2[col] = pd.to_numeric(df2[col], errors='coerce')

# === Agrupación CORRECTA ===
df2 = (
    df2
    .groupby(keys, as_index=False)[cols_num]
    .sum()
)

# df2.info()
# display(df2)

# Quitar columnas duplicadas que NO deben agregarse
cols_excluir = ['departamento', 'codDepartamento', 'ano']   # ajusta según tus coincidencias reales
df2 = df2.drop(columns=[c for c in cols_excluir if c in df2.columns])

# === Unir con el consolidado ===
df_consolidado_produccion = df_consolidado_produccion.merge(
    df2,
    on=keys,
    how='left',
    suffixes=('_base', '_df2')
)

# df_consolidado_produccion.info()
# display(df_consolidado_produccion)
```
